```python
#try to write out all the formulas first
from numpy import random as rd
import plotly.graph_objs as go

#Allele frequencies


#Fitnesses
fAA = 1
fAa = 1
faa = 1

#Pop Size
pop = 1000

#Number of generations
gen = 100

#Number of simulations
sim = 100
#for loop starts here
simulation = {}

#for histogram
finalfreqA = []

for i in range(sim+1):
    p0 = 0.50 #initA
    q0 = 0.50 #inita
    freqA = []
    freqA.append(p0)
    for j in range(gen+1):
        aliveAA = (p0**2)*fAA
        aliveAa = (2*p0*q0)*fAa
        aliveaa = (q0**2)*faa

        wbar = aliveAA+aliveAa+aliveaa
        freqAA = aliveAA/wbar
        freqAa = aliveAa/wbar
        freqaa = aliveaa/wbar

        newAA = rd.binomial(pop, freqAA, size=1)[0]
        newAa = rd.binomial(pop, freqAa, size=1)[0]
        newaa = pop-newAA-newAa
        #new population distribution with finite population size is generated

        p0 = (newAA + 0.5*newAa)/pop
        q0 = (newaa + 0.5*newAa)/pop
        freqA.append(p0)

        if j == gen:
            finalfreqA.append(p0)
        else:
            continue

    simulation[i] = freqA

#print (simulation)
#print(finalfreqA)

traces = []
for i in simulation.keys():
    plot_trace = go.Scatter(x = list(range(gen+1)), y = simulation[i], name=i)
    traces.append(plot_trace)

fig = go.Figure([i for i in traces])
fig.update_layout(yaxis_range=[0,1])
fig.show()



hist = go.Histogram(x = finalfreqA)
fig = go.Figure([hist])
fig.update_layout(xaxis_range=[0,1])
fig.show()

```
