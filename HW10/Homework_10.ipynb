```python
%%sql

SELECT title,CAI,seq FROM cai, sequences, orthologs WHERE CAI > 0.8
AND (gene_name = cerevisiae OR gene_name = mikatae OR gene_name = paradoxus OR gene_name = bayanus)
AND gene_name = title
```

__Exercise 4.__ Now generate a table with the number of genes per chromosome that have CAI larger than 0.5 and have orthologs in the other three yeast species.
