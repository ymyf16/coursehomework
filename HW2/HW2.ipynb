* Saccharin and sucralose impair glucose tolerance in healthy adults.

* Microbiome features correlate with NNS impacts on the human glycemic response.
* Causative personalized impacts of NNS-modulated microbiome on glycemic responses.
* Microbiome features correlate with person-specific variations in glycemic responses to NNS.
